```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

#Because my browser was facing issues with running the scrapper on this site, Yip(TA) helped me with an HTML file of the webpage and that's how I was able to scrap it.

doc = BeautifulSoup(open('driver-license-locations.html', 'r').read())
doc
```

```python
items = doc.find_all('div', attrs={'class': 'table-responsive'})[0].find_all('tr')

len(items)
```

```python
rows = []
#I struggled with this part as well, and it was due to a TA's help that I managed to successfully complete this.
for item in items:
    row = {}
    print(item.find_all('td')[0])
    print(item.find_all('td')[1])
    print(item.find_all('td')[2])
    print(item.find_all('td')[3])
```

```python
rows = []
#For this part, I took help from ChatGPT, as well as, a TA.
for item in items:
    cols = item.find_all('td')
    company_link = cols[0].find('a')
    url = company_link['href'] if company_link and company_link.has_attr('href') else 'no url'
    row = {
        "Company": cols[0].get_text(),
        "URL": url,
        "Address": cols[1].get_text(strip = True),
        "Telephone": cols[2].get_text(strip = True),            
        "Hours": cols[3].get_text(strip = True)
        }
    rows.append(row)
rows 
```

```python
import pandas as pd

df = pd.json_normalize(rows)
df.head(41)
```

```python
df.to_csv("Authorized_Third_Party_Driver_License_Locations.csv")
```
