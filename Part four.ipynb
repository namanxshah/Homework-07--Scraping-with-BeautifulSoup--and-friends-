Import all the libraries

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

Searching for 'car' gives paginated results with each page showing just 10 results. Since there are 132 results in total, I change the portion of url from 'zoom_per_page=10' to 'zoom_per_page=1000'. This expands the page results to show all the results in a single page.

```python
response = requests.get("https://www.tnwb.uscourts.gov/Search/search.aspx?zoom_query=car&zoom_page=1&zoom_per_page=1000&zoom_and=1&zoom_sort=0&zoom_xml=0")
doc = BeautifulSoup(response.text)
```

Inspecting the page tells that each car case is listed under 'result_block' and 'result_altblock'. So we extract them in a variable and check the total count.

```python
items = doc.find_all('div', attrs={'class':['result_block' , 'result_altblock']})
len(items)
```

Now, through the title of the case, we check if it is printing all the titles with a for loop.

```python
for item in items:
    print("------")
    print(item.find(class_='result_title').text)
```

```python
rows = []

for item in items:
    row = {}
    row['url'] = item.find('a')['href']
    row['case_name'] = item.find(class_='result_title').find('a').text
    row['category'] = item.find(class_='category').text.strip()
    row['other_details'] = item.find(class_='infoline').text
    rows.append(row)

df=pd.json_normalize(rows)
df.head()
```

We put them in separate rows, as asked in the homework question, and make a table. Then, check the table with df.head()
